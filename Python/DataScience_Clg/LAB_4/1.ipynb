# Lab Objective:
1. To simulate a population distribution (e.g. uniform or exponential distribution).
2. To take multiple random samples of a given size from this population and compute their means.
3. To observe how the distribution of these sample means approaches a normal distribution as the number of
males and sample size increase.

# Required Libraries:
1. numpy: For data manipulation and random sampling. 10
2. matplotlib: For visualization.
3. seaborn: For enhanced visualization.

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Set random seed for reproducibility
np.random.seed(42)
# Step 1: Define a population distribution (non-normal distribution)
population_size = 100000
sample_sizes = [2, 5, 10, 30, 50, 200] 
# Sample sizes to demonstrate convergent
num_samples = 1000 # Number of samples for each sample size

population = np.random.exponential(scale=2, size=population_size)
#Here, population_size - 100000, this will generate 100,000 random samples from 
print(population)
population_mean=np.mean(population)
print(population_mean)
```

```python
# Exponential distribution
plt.figure(figsize=(10, 6))
sns.histplot(population, bins=50, kde=True, color="green", alpha=0.7)  
plt.title("Population Distribution (Exponential)", fontsize=16)
plt.xlabel("Value", fontsize=14)  
plt.ylabel("Frequency", fontsize=14)  
plt.show()
```

The Central Limit Theorem Says:
+ If you take many random samples of a certain size from any population (doesn't matter if the population data is normal or not), and calculate the average (mean) of each sample:
+ The distribution of those sample averages will start to look like a bell-shaped curve (a normal distribution) as the sample size gets larger.

```python
plt.figure(figsize=(14, 10))

for i, size in enumerate(sample_sizes, start=1):
    sample_means = [np.mean(np.random.choice(population, size=size, replace=False)) for _ in range(num_samples)]
    # print(i)
    # print(size)
    # print(sample_means)
    # Plot the distribution of sample means
    plt.subplot(3, 2, i)
    sns.histplot(sample_means, bins=30, kde=True, color="orange", alpha=0.7)
    plt.title(f"Sample Means Distribution (n={size})", fontsize=14)
    plt.xlabel("Sample Mean", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)

# Step 4: Show the final plots 
plt.suptitle("Demonstration of the central limit theorem", fontsize=18,y=1.02)
plt.tight_layout()
plt.show()
```
