```python
# Data Preparation for heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
data = pd.read_csv('Titanic.csv')

# Select relevant numerical features
numeric_data = data[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]

# Handle missing values by filling with the median for better visaulization
numeric_data['Age'].fillna(numeric_data['Age'].median(), inplace=True)

print(numeric_data.head())
```

```python
plt.figure(figsize=(10, 6))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt='.2f', linewidths=0.5)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=data, palette='Set3')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
